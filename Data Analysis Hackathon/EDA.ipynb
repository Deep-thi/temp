```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
```

```python
df = pd.read_csv('imdb_top_1000.csv')
df.head()
```

```python
df.info()
```

```python
from collections import Counter
genre=[]
for x in df['Genre']:
    for y in x.split(','):
        genre.append(y.strip().lower())

count=Counter(genre)
count=count.most_common()[:10]
x,y=map(list,zip(*count))

fig,axs=plt.subplots(figsize=(20,5))
g=sns.barplot(y,x)
g.set_ylabel("Genres", weight = "bold")
g.set_title("Top Ten Genres", weight = "bold")
plt.show()
```

```python
fig,axs=plt.subplots(figsize=(20,5))
g=sns.kdeplot(df['duration'])
g.set_title("Time Duration of movies", weight = "bold")
```

```python
import re
duration=[]
for x in df["Runtime"]:
    p=re.compile('\d+')
    j=p.findall(x)
    if len(j)==2:
        duration.append(int(j[1])-int(j[0])+1)
    else:
        duration.append(int(j[0])+1)
df['duration']=duration
```

```python
top_voted = df.sort_values(['No_of_Votes'], ascending = False)
```

```python
df['Certificate']=df['Certificate'].fillna('Not Rated')

certificates=df[df['Certificate']!='Not Rated']['Certificate']
fig,axs=plt.subplots(figsize=(20,5))
g=sns.countplot(certificates)
g.set_title("Count of Certificates", weight = "bold")
plt.show()
```

```python
stars=['Star1','Star2','Star3','Star4']
fig,axs=plt.subplots(4,1,figsize=(20,7))
ax=0
for x in stars:
    axs[ax].bar(df[x].value_counts()[:10].index,df[x].value_counts()[:10])
    axs[ax].set_title(x)
    axs[ax].set_ylabel("Appearances", weight = "bold")
    ax+=1
    plt.tight_layout()
```

```python
stars=['Star1','Star2','Star3','Star4']
fig,axs=plt.subplots(4,1,figsize=(20,7))
ax=0
for x in stars:
    s=df.groupby([x]).sum().reset_index()
    d=s.sort_values(['No_of_Votes'],ascending=False)[:10]
    axs[ax].bar(d[x],d['No_of_Votes'])
    axs[ax].set_title(x)
    axs[ax].set_ylabel("Appearances", weight = "bold")
    ax+=1
    plt.tight_layout()
```
